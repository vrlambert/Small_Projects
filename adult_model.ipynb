# Modeling the adult dataset

```python
import pandas as pd
import matplotlib
import seaborn
```

```python
header_names = ['age',
        'workclass',
        'fnlwgt',
        'education',
        'education_num',
        'marital_status',
        'occupation',
        'relationship',
        'race',
        'sex',
        'capital_gain',
        'capital_loss',
        'hours_per_week',
        'native_country',
         'salary_label']
train = pd.read_csv('adult.data', 
                    header = None, 
                    names = header_names, 
                    index_col = False)
train.head()
```

```python
test = pd.read_csv('adult.test', 
                   header = None, 
                   names = header_names, 
                   index_col = False, 
                   skiprows = 1)
test.head()
```
